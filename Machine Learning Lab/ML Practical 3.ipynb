```python
import matplotlib.pyplot as plt
import numpy as np
arr = np.random.rand(500)
_ = plt.hist(arr, bins=10, rwidth=0.9)
plt.xlabel('Random Values')
plt.ylabel('Occurences')
```

```python
plt.plot(np.sin(np.arange(0, 10, step=0.1)))
plt.title('Sine/Cosine Plot')
plt.plot(np.cos(np.arange(0, 10, step=0.1)))
plt.legend(['Sine', 'Cosine'])
```
